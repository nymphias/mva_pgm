```python
%load_ext autoreload
%autoreload 2
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline

np.random.seed(1)

import tools.Dataset as dt
import models.KMeansClustering as KMeans
import models.EMGaussian as EMG
```

```python
data = dt.Dataset()
```

# K-means clustering

```python
for i in range(5):
    model = KMeans.KMeansClustering(n_clusters=4, max_iter=300, tol=1e-5, init_mode = "kmpp")
    model.fit(data.X_train)
    model.plot_clusters()
```

```python
model = EMG.EMGaussian(n_clusters=4, max_iter=300, tol=1e-3, init_mode = "random", ellipse_level=0.90)
# model = EMG.EMGaussian(n_clusters=4, max_iter=300, tol=1e-3, init_mode = "random", nstd=np.sqrt(5.991))
model.fit(data.X_train)
model.plot_clusters()
```

```python
model = EMG.EMGaussian(n_clusters=4, max_iter=300, tol=1e-3, init_mode = "random", ellipse_level=0.90)
# model = EMG.EMGaussian(n_clusters=4, max_iter=300, tol=1e-3, init_mode = "random", nstd=np.sqrt(5.991))
model.fit(data.X_train)
```

```python
labels = model.predict(data.X_test)
```

```python
plt.scatter(data.X_test[:, 0], data.X_test[:, 1], c=labels, cmap=plt.get_cmap('viridis'))
plt.scatter(model.cluster_centers_[:, 0], model.cluster_centers_[:, 1], c='b', label='Final Centers', marker='v')

plt.suptitle(f'[n_clusters={model.n_clusters_}] GMM on test data', size=14)
plt.title(f'-LogLikelihood = {model.loss_history_[model.n_iter_ - 1]}')
plt.legend()
plt.show();
```

```python
plt.scatter(data.X_train[:, 0], data.X_train[:, 1], c=model.labels_, cmap=plt.get_cmap('viridis'))
plt.title("X_train")
plt.show()
plt.scatter(data.X_test[:, 0], data.X_test[:, 1], c=labels, cmap=plt.get_cmap('viridis'))
plt.title("X_test")
plt.show()
```
