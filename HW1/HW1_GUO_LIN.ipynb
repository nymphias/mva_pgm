```python
%load_ext autoreload
from models import LinReg as linr
from models import LogisticRegression as logr
from models import LDA as LDA
from models import QDA as QDA


import numpy as np
import pandas as pd
%matplotlib inline

import tools.Dataset as dt
import matplotlib.pyplot as plt

%autoreload 2
```

```python
dataA, dataB, dataC = dt.Dataset('A'), dt.Dataset('B'), dt.Dataset('C')
```

```python
data = dataA
```

# LDA

```python
model = LDA.LDA()
model.fit(X=data.X_train, y=data.y_train)

data.y_train_pred['LDA'] = model.predict(data.X_train)
data.y_test_pred['LDA'] = model.predict(data.X_test)

data.error_train['LDA'] = dt.misclassification_error(data.y_train, data.y_train_pred['LDA'])
data.error_test['LDA'] = dt.misclassification_error(data.y_test, data.y_test_pred['LDA'])

print(f"Train misclassification = {data.error_train['LDA']}")
print(f"Test misclassification = {data.error_test['LDA']}")

################################# PLOTTING ##############################
# Calculate P(y=1|x) = 0.5 for LDA

x1_list = np.arange(-10, 10, 0.1)
x2_list = np.arange(-10, 10, 0.1)
x1, x2 = np.meshgrid(x1_list, x2_list)
values = model._compute_contour(np.array([x1.flatten(), x2.flatten()]).T).reshape((len(x1_list), len(x2_list)))

# train data
label_one = data.y_train == 1
plt.figure()
plt.scatter(data.X_train[label_one, 0], data.X_train[label_one, 1], label='1')
plt.scatter(data.X_train[~label_one, 0], data.X_train[~label_one, 1], label='0')
plt.contour(x1, x2, values, [0], label = 'Boundary')
plt.legend()
plt.title(f'[dataset={data.dataset}] train data')
plt.suptitle(f'{model.name}')
plt.show()

# test data
label_one = data.y_test == 1
plt.figure()
plt.scatter(data.X_test[label_one, 0], data.X_test[label_one, 1], marker = '.', s = 20, label='1')
plt.scatter(data.X_test[~label_one, 0], data.X_test[~label_one, 1], marker = 'x', s = 20,label='0')
plt.contour(x1, x2, values, [0], label = 'Boundary')

plt.legend()
plt.title(f'[dataset={data.dataset}] test data')
plt.suptitle(f'{model.name}')
plt.show()

del x1,x2
```
